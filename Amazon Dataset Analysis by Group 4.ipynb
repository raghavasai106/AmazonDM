```python
import plotly.graph_objects as go
import pandas as pd
e = pd.read_csv("Amazon_Dataset.csv")
df=pd.DataFrame(e)

fig = go.Figure([go.Histogram(x=df['category'], y=df['manufacturer'].unique())])
fig.update_layout(title_text="Number of manufacturers for each category  ")
fig.update_xaxes(title_text="Category")
fig.update_yaxes(title_text="Number of manufacturer")
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

e = pd.read_csv ('/Users/raghavasaikowlasker/PycharmProjects/AmazonDM/Amazon_Dataset.csv')
df=pd.DataFrame(e)
filter_reviews = df[df['number_of_reviews'] > 300]
fig = px.pie(filter_reviews, values='number_of_reviews', names='manufacturer',title='Total Reviews for Each Manufacturer',color_discrete_sequence=px.colors.sequential.RdBu
             , hover_data=['category'], labels={'reviews':'number_of_reviews'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
e = pd.read_csv("Amazon_Dataset.csv")
df=pd.DataFrame(e)
fig = px.histogram(df, y= df['Rating'], x="category", color_discrete_sequence=['indianred'])
fig.update_layout(title_text="Highest gross Ratings for Each Category")
fig.update_xaxes(title_text="Category")
fig.update_yaxes(title_text="Total Ratings")
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
e = pd.read_csv("Amazon_Dataset.csv")
df=pd.DataFrame(e)

fig = px.pie(df, values='Rating', names='category',title='Gross Ratings for each Category'
             , hover_data=['category'], labels={'reviews':'number_of_reviews'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()

```
