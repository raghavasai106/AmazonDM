```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
e = pd.read_csv("new_amazon_data.csv")
df=pd.DataFrame(e)

fig = go.Figure([go.Bar(x=df['category'], y=df['Rating'])])
fig.update_layout(title_text="Highest gross ratings for each category")
fig.show()



# This dataframe has 244 lines, but 4 distinct values for `day`
# fig = px.pie(df.head(1000), values='Rating', names='category', title = 'Highest ratings for each category')
# fig.show()

```

```python
import plotly.express as px
import math
e = pd.read_csv("new_amazon_data.csv")
df=pd.DataFrame(e)
#df = px.data.gapminder(e)

fig = px.scatter(df, x="price", y="Rating", size="Rating"
,color="price", hover_name="product_name_new", log_x=True, size_max=30, title = "")
fig.update_layout(title_text="Price of Each Product with respect to its Rating")
fig.update_traces(mode='markers', marker=dict(sizemode='area',
                                               line_width=2))
fig.show()
```
